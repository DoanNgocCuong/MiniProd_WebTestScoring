### *1. WHY? MỤC ĐÍCH LÀM, TẠI SAO LÀM, VIỆC NÀY CÓ QUAN TRỌNG KHÔNG, CẦN THIẾT KHÔNG?*
- Mục đích chính của việc xây dựng bot là gì? (Ví dụ: nâng cao kiến thức sản phẩm, cải thiện kỹ năng bán hàng, đánh giá năng lực hiện tại của nhân viên)
- Tại sao việc xây dựng bot lại quan trọng? (Ví dụ: tăng cường hiệu quả đào tạo, tiết kiệm thời gian và chi phí, cải thiện chất lượng bán hàng)
- Việc này có cần thiết không? (Ví dụ: Có yêu cầu cụ thể từ quản lý, nhân viên cần công cụ hỗ trợ, nhu cầu thị trường)

### *2. WHO? TRÊN THỊ TRƯỜNG CÓ CON BOT NÀO NHƯ THẾ CHƯA?*
- Chị có con bot mẫu nào mà chị thấy ưng không? (Em có quét nhanh thị trường và thấy một số con bot như: [Tên bot 1], [Tên bot 2], [Tên bot 3]... Các bot này có những chức năng như đào tạo kiến thức sản phẩm, kiểm tra kỹ năng bán hàng, và phân tích phản hồi từ nhân viên)
- Chị đã từng sử dụng hoặc trải nghiệm bot nào tương tự chưa? (Nếu có, chị có thể chia sẻ trải nghiệm và những điểm chị thấy tốt và cần cải thiện không?)
- Trên thị trường hiện nay, bot nào đang được đánh giá cao nhất về hiệu quả đào tạo và kiểm tra kỹ năng bán hàng? (Ví dụ: [Tên bot], điểm nổi bật của các bot này)
- Chị có biết những điểm mạnh và điểm yếu của các bot hiện có trên thị trường không? (Nếu có, chị có thể chia sẻ chi tiết không?)
- Chị có mong muốn bot của chúng ta có điểm gì khác biệt so với các bot hiện tại trên thị trường không? (Ví dụ: tính năng mới, cải thiện UX/UI, tích hợp công nghệ mới)

### **3. WHAT? LÀM RÕ REQUIREMENTS
 *1. Yêu cầu tổng quan *
- Mục tiêu chính của việc xây dựng bot là gì? (Ví dụ: nâng cao kiến thức sản phẩm, cải thiện kỹ năng bán hàng, đánh giá năng lực hiện tại của nhân viên)
- Đối tượng sử dụng bot là ai? (Ví dụ: nhân viên mới, nhân viên hiện tại, quản lý)
- Bot sẽ hoạt động trên nền tảng nào? (Ví dụ: GPTs, web, mobile, chat application). Em có từng xài nền tảng Mindmaid, ... chị có muốn xem thử. 


 *2. Yêu cầu chức năng*
- Bot cần có những chức năng gì? (Ví dụ: bài đào tạo theo ngày, quiz/trắc nghiệm, mô phỏng cuộc gọi bán hàng, đưa ra feedback cho nhân viên, ... )
- Bot sẽ quản lý và cung cấp các tài liệu giới thiệu sản phẩm như thế nào? (Tích hợp với hệ thống quản lý tài liệu, cập nhật nội dung tự động, hay là cập nhật bằng cơm trên GPTs, ... Độ dài của các tài liệu này)
- Bot sẽ giúp đào tạo và kiểm tra các nghiệp vụ sales như thế nào? (Ví dụ: mô phỏng tình huống bán hàng, bài kiểm tra lý thuyết, thực hành kỹ năng)
- Bot cần có khả năng tự động cập nhật nội dung mới hay không? (Có hoặc không, dựa trên yêu cầu quản lý nội dung)

*3. Yêu cầu về nội dung*
- Những thông tin chi tiết nào về sản phẩm cần được cung cấp qua bot? (Thông tin sản phẩm chi tiết, tính năng, lợi ích, giá cả, thị trường mục tiêu)
- Những kiến thức nghiệp vụ sales nào cần được bao gồm trong chương trình đào tạo? (Kỹ năng bán hàng, quy trình bán hàng, kỹ năng giao tiếp, kỹ năng đàm phán)
- Có yêu cầu về ngôn ngữ hoặc văn phong nào đặc biệt không? (Ngôn ngữ formal, informal, technical)

*4. Yêu cầu về trải nghiệm người dùng*
- Giao diện và trải nghiệm người dùng mong muốn như thế nào? (Giao diện trực quan, dễ sử dụng, thân thiện với người dùng)
- Có yêu cầu gì về giao diện trực quan, dễ sử dụng không? (Giao diện đơn giản, dễ điều hướng, hỗ trợ đa ngôn ngữ)
- Bot cần có khả năng giao tiếp bằng giọng nói hay chỉ qua văn bản? (Cả hai hoặc chỉ qua văn bản, tùy theo nhu cầu sử dụng)
 *5. Yêu cầu kỹ thuật*
- Có yêu cầu cụ thể nào về công nghệ sử dụng để xây dựng bot không? (AI/ML frameworks, cloud services)
- Bot cần tích hợp với hệ thống nào khác không? (CRM, LMS)
- Có yêu cầu về bảo mật và quyền riêng tư dữ liệu không? (Các tiêu chuẩn bảo mật, GDPR, quyền truy cập dữ liệu)

 *6. Yêu cầu về quản lý và vận hành*
- Ai sẽ chịu trách nhiệm quản lý và cập nhật nội dung cho bot? (Đội ngũ quản lý nội dung, nhân viên IT)
- Cần có các báo cáo và phân tích gì từ hoạt động của bot? (Kết quả kiểm tra, tiến độ đào tạo, phản hồi người dùng)
- Có yêu cầu gì về hỗ trợ và bảo trì bot sau khi triển khai không? (Hỗ trợ kỹ thuật, cập nhật định kỳ, bảo trì hệ thống)

 *7. Yêu cầu về hiệu quả và đánh giá*
- Làm thế nào để đánh giá hiệu quả của bot trong việc đào tạo và kiểm tra nhân viên sales? (Kết quả kiểm tra, phản hồi của nhân viên, cải thiện hiệu suất bán hàng)
- Các tiêu chí nào sẽ được sử dụng để đo lường thành công? (Kết quả kiểm tra, tỷ lệ hoàn thành đào tạo, mức độ hài lòng của nhân viên)

 *8. Yêu cầu khác*
- Có bất kỳ yêu cầu hoặc mong muốn đặc biệt nào khác từ phía bạn hoặc các bên liên quan không? (Các yêu cầu cụ thể khác từ khách hàng hoặc quản lý)

### *4. OKRs (Objectives and Key Results)*
- Mục tiêu cụ thể của dự án là gì? (Ví dụ: hoàn thành bot trong 3 tháng, tăng hiệu suất bán hàng lên 20% sau 6 tháng)
- Các kết quả chính cần đạt được là gì? (Ví dụ: số lượng nhân viên hoàn thành khóa đào tạo, tỷ lệ nhân viên đạt kết quả kiểm tra tốt, mức độ hài lòng của nhân viên với bot)

### *5. FEEDBACKs*
- Bạn sẽ thu thập phản hồi từ người dùng như thế nào? (Ví dụ: khảo sát trực tiếp, đánh giá qua bot, họp định kỳ)
- Làm thế nào để cải thiện bot dựa trên phản hồi của người dùng? (Ví dụ: cập nhật nội dung, cải thiện giao diện, sửa lỗi kỹ thuật)
- Có yêu cầu cụ thể gì về việc theo dõi và phân tích phản hồi không? (Ví dụ: công cụ phân tích, báo cáo định kỳ)

### *6. Dự đoán những tình huống ngoại lệ*
- Tình huống nhân viên không có kỹ năng công nghệ cần thiết để sử dụng bot (Giải pháp: Cung cấp tài liệu hướng dẫn sử dụng chi tiết, tổ chức buổi đào tạo sử dụng bot)
- Tình huống nội dung đào tạo cần cập nhật liên tục (Giải pháp: Bot cần có khả năng tự động cập nhật nội dung mới, hoặc dễ dàng cập nhật qua hệ thống quản lý nội dung)
- Tình huống nhân viên không có thời gian để hoàn thành các bài kiểm tra (Giải pháp: Cho phép nhân viên hoàn thành các bài kiểm tra vào thời gian rảnh, cung cấp thông báo nhắc nhở)
- Tình huống bot gặp sự cố kỹ thuật (Giải pháp: Cần có đội ngũ hỗ trợ kỹ thuật sẵn sàng xử lý sự cố, hệ thống sao lưu và khôi phục dữ liệu)
- Tình huống nội dung không phù hợp hoặc không hiệu quả (Giải pháp: Thu thập phản hồi người dùng để cải thiện nội dung, thực hiện các bài kiểm tra đánh giá hiệu quả định kỳ)

- Bắc cầu - Nguyên tắc Giàn giáo: Chuyển đổi từ kiến thức sang kỹ năng theo từng bước, có khung cấu trúc và hệ thống rõ ràng. Độ phức tạp được giới thiệu dần dần, tạo ra một quá trình học tập tích lũy mạnh mẽ.
