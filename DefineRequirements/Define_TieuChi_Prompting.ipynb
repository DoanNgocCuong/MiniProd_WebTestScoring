#### Step 1: Giải thích ngắn gọn 7 tiêu chí

1. Rõ ràng và dễ hiểu: Câu hỏi được diễn đạt rõ ràng, không gây hiểu lầm hay mơ hồ cho người trả lời.
2. Liên quan và phản ánh đúng nội dung đào tạo: Câu hỏi tập trung vào những nội dung đã được giảng dạy trong quá trình training và kiểm tra kiến thức liên quan.
3. Thách thức và phù hợp với mức độ kỹ năng của người học: Câu hỏi cần có mức độ khó hợp lý, không quá dễ để đảm bảo thách thức người học nhưng cũng không quá khó để tránh gây nản lòng.
4. Kiểm tra được khả năng áp dụng thực tế của kiến thức: Câu hỏi nên yêu cầu người học vận dụng kiến thức đã học vào các tình huống thực tế trong công việc.
5. Đa dạng và toàn diện, đánh giá nhiều khía cạnh của kiến thức và kỹ năng: Bộ câu hỏi cần bao gồm nhiều dạng câu hỏi khác nhau để kiểm tra toàn diện các kỹ năng và kiến thức.
6. Có thể đo lường và đánh giá được hiệu quả của quá trình training: Câu hỏi cần có thang điểm rõ ràng, cho phép dễ dàng đo lường mức độ hiểu biết và kỹ năng của người học.
7. Tạo cơ hội cho người học tự đánh giá và phản ánh: Câu hỏi nên kích thích người học tự đánh giá quá trình học tập của mình và phản ánh về cách họ áp dụng kiến thức đã học.
