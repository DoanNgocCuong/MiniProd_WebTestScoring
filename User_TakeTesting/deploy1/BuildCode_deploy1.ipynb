4.1 Gradio có các nút nào?

 Input Controls

- Buttons: Các nút để người dùng thực hiện các hành động như submit, reset, clear, etc.[1][2]
- Text Inputs: Các ô nhập liệu cho phép người dùng nhập văn bản[1][2]
- Dropdowns: Các menu thả xuống để người dùng chọn một giá trị từ danh sách[1][2]
- Checkboxes: Các ô chọn cho phép người dùng chọn nhiều tùy chọn[1][2]
- Radio Buttons: Các nút radio để người dùng chọn một tùy chọn duy nhất[1][2]
- Sliders: Các thanh trượt để người dùng chọn một giá trị trong một khoảng[1][2]

 Navigational Components

- Tabs: Các tab để người dùng chuyển đổi giữa các nội dung khác nhau[1][2]
- Breadcrumbs: Các liên kết giúp người dùng xem vị trí hiện tại và di chuyển lên các cấp trên[2]
- Pagination: Các nút và số trang để người dùng chuyển đổi giữa các trang[2]

 Informational Components

- Tooltips: Các chú thích hiện lên khi hover chuột vào một phần tử[1][2]
- Modals: Các cửa sổ popup hiện lên trên giao diện chính[1][2]
- Progress Bars: Các thanh tiến trình để hiển thị quá trình hoàn thành của một tác vụ[1][2]
- Notifications: Các thông báo hiện lên để cập nhật người dùng về tình trạng của một hành động[1][2]

#### 1.1 UI Trắc Nghiệm

Build UI with Gradio 
1. Giao diện bắt đầu: có nút `Start`
2. Giao diện khi đã nhấn `Start`:
    - Ô `response_and_question`: 
        - Ban đầu câu hỏi 1 thì `response` rỗng => UI hiển thị `response_and_question` là câu hỏi 1
        - Sau khi ấn Enter thì màn hình hiển thị ô `response_and_question` là: response của câu 1 và question câu 2 và ô `user_input` để input câu trả lời của user
        - User nhập đáp án vào ô `user_input` và ấn Enter => UI hiển thị ô `response_and_question` là response câu 2 và question câu 3 và ô `user_input` để input câu trả lời của user
    - ô `user_input` ô để input câu trả lời của user

    - Khi hiển thị "Tổng điểm của bạn là : ..." tức là đã hết câu hỏi, thì ko hiển thị nút `user_input` nữa 
    - Chỉ hiển thị nút "Start" 1 lần lúc đầu. (    - Ẩn nút "Start" sau khi đã hiển thị response_and_question)

````python
import pandas as pd
import gradio as gr

# Backend Function
def load_quiz_data(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    return df['question'].tolist(), df['checking_answer'].tolist(), df['explain_answer'].tolist()

def quiz_logic(questions, checking_answers, explain_answers, user_answer=None, current_question_index=0, total_score=0):
    if current_question_index >= len(questions) or pd.isna(questions[current_question_index]):
        return f"Tổng điểm của bạn là: {total_score}/{len(questions)}", gr.update(visible=False), gr.update(visible=False)
    
    if user_answer:
        if user_answer.strip().upper() == checking_answers[current_question_index]:
            assistant_response = f"Đúng! Get 1 point.\n{explain_answers[current_question_index]}"
            total_score += 1
        else:
            assistant_response = f"Sai! Get 0 point.\n{explain_answers[current_question_index]}"
        
        response_and_question = (
            f"YOUR ANSWER: {user_answer}\n"
            f"FEEDBACK: \n{assistant_response}\n"
            f"---------------------"
        )
        
        current_question_index += 1
    else:
        response_and_question = ""
    
    if current_question_index < len(questions) and not pd.isna(questions[current_question_index]):
        response_and_question += f"\nCâu hỏi: {questions[current_question_index]}"
        return response_and_question, gr.update(visible=True), gr.update(visible=True), current_question_index, total_score
    else:
        response_and_question += f"\nTổng điểm của bạn là: {total_score}/{len(questions)}"
        return response_and_question, gr.update(visible=True), gr.update(visible=False), current_question_index, total_score

# Frontend Function
def create_quiz_interface(file_path, sheet_name):
    questions, checking_answers, explain_answers = load_quiz_data(file_path, sheet_name)
    total_score = 0
    current_question_index = 0

    with gr.Blocks() as demo:
        start_button = gr.Button("Start")
        response_and_question = gr.Textbox(label="Response and Question", interactive=False, visible=False)
        user_input = gr.Textbox(label="Your Answer (A/B/C/D)", visible=False)
        
        def start_quiz():
            nonlocal current_question_index, total_score
            quiz_result, response_visibility, user_input_visibility, current_question_index, total_score = quiz_logic(
                questions, checking_answers, explain_answers, current_question_index=current_question_index, total_score=total_score
            )
            return quiz_result, gr.update(visible=False), response_visibility, user_input_visibility

        start_button.click(start_quiz, inputs=None, outputs=[response_and_question, start_button, response_and_question, user_input])
        
        def handle_user_input(user_answer):
            nonlocal current_question_index, total_score
            quiz_result, response_visibility, user_input_visibility, current_question_index, total_score = quiz_logic(
                questions, checking_answers, explain_answers, user_answer, current_question_index, total_score
            )
            return quiz_result, response_visibility, user_input_visibility

        user_input.submit(handle_user_input, inputs=user_input, outputs=[response_and_question, response_and_question, user_input])

    demo.launch()

# File path and sheet name
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\TASK5_GPTschatbot_TestingSale\Input\[SaleViewer]Define_bài_toán_SalerTesting.xlsx'
sheet_name = 'TN_P1P2'

# Create and launch the quiz interface
create_quiz_interface(file_path, sheet_name)
````

````python
import pandas as pd
from openai import OpenAI
import gradio as gr
import json

def scoring_short_essay_questions(QUESTION, CHECKING_ANSWER, USER_ANSWER):
    client = OpenAI(api_key="")

    prompt = f"""
Act: You are an expert at scoring short essay questions. 
You are provided with 1 QUESTION and 1 CHECKING ANSWER for that QUESTION. After USER enters the answer, you will check USER ANSWER with CHECKING ANSWER. 

Instructions: 
- point: check USER ANSWER with CHECKING ANSWER to score on a 3-point scale 
0 points: The answer is completely unrelated to the question or has no content.
1 point: The answer may have a few relevant keywords but does not clarify the main idea at all (or lacks a connection).
2 points: The answer is correct but lacks analysis (or has analysis but is still sketchy).
3 points: The answer is correct, has clear analysis, and strong arguments. 

- comment: briefly in 2 sentences: the reason you gave such a score (in Vietnamese)
- answer_3_point: correct the user's original sentence to gain 3 points

Output: JSON format with 2 keys: `point`, `comment`, and `answer_3_point`

QUESTION: 
{QUESTION}

CHECKING ANSWER: 
{CHECKING_ANSWER}
"""

    completion = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": USER_ANSWER},
        ], 
        temperature=0, 
        max_tokens=6000,
    )

    return completion.choices[0].message.content

def run_quiz(file_path, sheet_name):
    # Load the Excel file and the specific sheet
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # Extract relevant columns
    questions = df['question']
    checking_answers = df['checking_answer']

    # Initialize score
    total_score = 0
    current_question_index = 0

    def quiz_logic(user_answer=None):
        nonlocal current_question_index, total_score

        if current_question_index >= len(questions) or pd.isna(questions[current_question_index]):
            return f"Tổng điểm của bạn là: {total_score}/{len(questions)}", gr.update(visible=False), gr.update(visible=False)

        if user_answer:
            response = scoring_short_essay_questions(questions[current_question_index], checking_answers[current_question_index], user_answer)
            response_json = json.loads(response)
            point = response_json['point']
            comment = response_json['comment']
            answer_3_point = response_json['answer_3_point']

            total_score += point

            response_and_question = (
                f"- YOUR ANSWER: {user_answer}\n"
                f"- GET POINT: {point}/3 \n"
                f"- FEEDBACK: \n{comment}\n"
                f"- CORRECTED ANSWER: \n{answer_3_point}\n"
                f"---------------------"
            )

            current_question_index += 1
        else:
            response_and_question = ""

        if current_question_index < len(questions) and not pd.isna(questions[current_question_index]):
            response_and_question += f"\nCâu hỏi: {questions[current_question_index]}"
            return response_and_question, gr.update(visible=True), gr.update(visible=True)
        else:
            response_and_question += f"\nTổng điểm của bạn là: {total_score}/{len(questions)*3}"  # Mỗi câu 3 điểm
            return response_and_question, gr.update(visible=True), gr.update(visible=False)

    with gr.Blocks() as demo:
        start_button = gr.Button("Start")
        response_and_question = gr.Textbox(label="Response and Question", interactive=False, visible=False)
        user_input = gr.Textbox(label="Your Answer", visible=False)

        def start_quiz():
            quiz_result, response_visibility, user_input_visibility = quiz_logic()
            return quiz_result, gr.update(visible=False), response_visibility, user_input_visibility

        start_button.click(start_quiz, inputs=None, outputs=[response_and_question, start_button, response_and_question, user_input])

        def handle_user_input(user_answer):
            quiz_result, response_visibility, user_input_visibility = quiz_logic(user_answer)
            return quiz_result, response_visibility, user_input_visibility

        user_input.submit(handle_user_input, inputs=user_input, outputs=[response_and_question, response_and_question, user_input])

    demo.launch()

# Example usage
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\TASK5_GPTschatbot_TestingSale\Input\[SaleViewer]Define_bài_toán_SalerTesting.xlsx'
sheet_name = 'TL_P1P2'
run_quiz(file_path, sheet_name)
````
